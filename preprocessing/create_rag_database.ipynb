```python
%pip install chromadb openai
import chromadb
from utils import *

# Chroma Client
client = chromadb.PersistentClient(path="../chromadb/")
clean_collection = client.get_or_create_collection("clean_articles")
adversarial_collection = client.get_or_create_collection("adversarial_articles")

# Load the clean and adversarial articles
combined_clean_articles = load_json('../datasets/combined_clean_data.json')
combined_adversarial_articles = load_json('../datasets/combined_adversarial_data.json')

combined_clean_articles_texts = [v for _, v in combined_clean_articles.items()]
combined_adversarial_articles_texts = [v for _, v in combined_adversarial_articles.items()]
clean_id = [f"clean{i+1}" for i in range(len(combined_clean_articles_texts))]
adversarial_id = [f"adversarial{i+1}" for i in range(len(combined_adversarial_articles_texts))]
```

```python
# Create RAG db with all the clean articles
clean_collection.add(
    documents = combined_clean_articles_texts,
    ids = clean_id,
)

# Create RAG db with all the adversarial articles
adversarial_collection.add(
    documents = combined_adversarial_articles_texts,
    ids = adversarial_id,
)
```

```python
# Test query clean
results = clean_collection.query(
    query_texts = ["George Washington"],
    n_results = 1
)

print(results['documents'][0][0])

# Test query adversarial
results = adversarial_collection.query(
    query_texts = ["Who won the 1906 Iowa gubernatorial election"],
    n_results = 1
)

print(results['documents'][0][0])
```
